### Dependencies

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Load Data

```python
df = pd.read_csv('./data/legend.csv')
del df['user.id']
```

### Show 5 Examples of Our Data

```python
stop_counter = 0
for idx, row in df.iterrows():
    if stop_counter < 5:
        stop_counter += 1
        img_path = './images/' + row['image']
        img = mpimg.imread(img_path)
        imgplot = plt.imshow(img)
        plt.show()
        print('Label:', row['emotion'])
    else:
        break
```

### Dataset Info

```python
dataset_size = len(df)
```

```python
expressions = []
freq_table = dict()
for val in df['emotion']:
    val = val.lower()
    expressions.append(val)
    
    if freq_table.get(val):
        freq_table[val] += 1
    else:
        freq_table[val] = 1
    
expressions = list(set(expressions))
```

```python
print('Images:', dataset_size)
print('Expression Types:', expressions)
```

### Distribution of Expressions in the Facial Expression Dataset

```python
font = {'weight':'bold','size':22}
matplotlib.rc('font', **font)
plt.figure(figsize=(20, 20))
plt.bar(freq_table.keys(), freq_table.values())
```
